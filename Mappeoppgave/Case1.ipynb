```python
# Funksjonen viser hvordan leieprisen varierer utifra hvor mange boliger tilgjengelig (x),
# når utleier ønsker å leie ut til ca. 8000 kr.
def tilbud2(x):
    a = 800000/(100+x)
    return a

# Funksjonen viser hvordan prisen man er villig til å betale utifra etterspørsel (x),
# når man i ugangspunktet ønsker å leie for ca. 7000 kr.

def etterspørsel2(x):
    a = 700000/(100-x)
    return a
```

## Oppgave 4 (20 poeng)

Et lån består som regel av et månedlig terminbeløp. Dette beløpet er summen av avdrag (nedbetalingen på lånet) og renter. Vi antar månedlig forrenting i alle oppgavene. Dvs. at det er 12 terminer i hvert år.<br><br>
a) Lag en funksjon som regner ut hvor mye lånet "x" koster deg i renteutgifter for "t" terminer med årlig rente "r" for et serielån. <br> <br>  Siden dette er et serielån, så vil avdragene være like hver måned men renteutgiftene reduseres i takt med avdragene. Renteutgiftene for en gitt termin "t" vil derfor være den årlige renten "r" (delt på antall forrentinger "f") på gjenværende beløp på det tidspunktet.   $renteutgifter_{t} = (x-a*t)*{r/f}$   <br>  <br> 
Siden vi er ute etter den totale kostnaden i svaret, må du summere renteutgiftene over alle terminer, det vil si $\sum_{t=1}^N(x-a*t)*{r/f}$


<br>
<b>Hint:</b> siden terminbeløpet varierer for hver måned (pga at rentene endres), må alle enkeltperioder summeres. Det kan være nyttige å bruke funksjonen np.arange() til dette.


```python
import numpy as np

#Inputs: x = lånesum, T = terminer, r = årlig rente, f =  antall forentninger årlig

def renteutgift(x, T, f, r):
    a = x / T
    renteutgifter = []
    
    for t in range(T): #For hver termin t opp til totale antall terminer T
        rente_t = (x - a*t)*r/f
        renteutgifter.append(rente_t) #append legger sammen listene.
    return sum(renteutgifter)

```

b) regn ut hvor mye lånet koster deg med henholdsvis 10, 20 og 30 års tilbakebetaling. Anta 1 000 000 kr lånebeløp med 3% rente

```python
print("Et lån på 1 000 000 kr med en rente på 3%, koster følgende med 10 års tilbakebetaling:")
print(renteutgift(1000000, 120, 12, 0.03))
print(" ")
print("Et lån på 1 000 000 kr med en rente på 3%, koster følgende med 20 års tilbakebetaling:")
print(renteutgift(1000000, 240, 12, 0.03))
print(" ")
print("Et lån på 1 000 000 kr med en rente på 3%, koster følgende med 30 års tilbakebetaling:")
print(renteutgift(1000000, 360, 12, 0.03))
```

c) Vis hva det samme lånet koster som annuitetslån, dvs differansen mellom alle terminbeløp og lånebeløp.<br> <br> Annuitetslån gir like terminbeløp hver måned, men renten utgjør en større del av dette beløpet i starten. Terminbeløpet for et annuitetslån er definert ved formelen:
$T = x*\frac{r/f} {(1-(1+(r/f))^{-t})}$, hvor x=lånebeløp, r = årlig rente, t = terminer, f= antall forrentinger





```python
def terminbeløp(x, r, t, f):
    return x*((r/f)/(1-(1+(r/f))**(-t)))

print("Et lån på 1 000 000 kr med en rente på 3%, koster følgende med 10 års tilbakebetaling:")
print(str(int(12*10*terminbeløp(1000000, 0.03, 120, 12)-1000000)) + " kr") 
print(" ")
print("Et lån på 1 000 000 kr med en rente på 3%, koster følgende med 20 års tilbakebetaling:")
print(str(int(12*20*terminbeløp(1000000, 0.03, 240, 12)-1000000)) + " kr") 
print(" ")
print("Et lån på 1 000 000 kr med en rente på 3%, koster følgende med 30 års tilbakebetaling:")
print(str(int(12*30*terminbeløp(1000000, 0.03, 360, 12)-1000000)) + " kr") 
```

Dvs, hvis t=15, har det gått 1 år og 3 mnd med terminer. Se forøvrig relevante formler i oppgave (a)

<b>Hint2:</b> Siden vi er ute etter både rentekostnader og avdrag hver for seg, kan det være lurt å definere en funksjon for hver av dem. </br>


```python
#Lager funkjoner.

def renteutgift2(x, T, f, r): #liste over nedbetalingene
    a = x / T
    renteutgifter = []
    
    for t in range(T): #For hver termin t opp til totale antall terminer T
        rente_t = (x - a*t)*r/f
        renteutgifter.append(rente_t)
    return renteutgifter

def avdrag(x, T): #avdrag
    a = x/T
    return a

def avdrag2(x, T): #i liste form
    b = []
    a = x/T
    b.append(a)
    return b
```

```python
import matplotlib.pyplot as plt #pakke pyplot

nedbetaling_av_renter =np.array(renteutgift2(1000000, 240, 12, 0.03)) #array av renteutgiftene
avdrag240 = avdrag2(1000000, 240) * 240 #lager en liste på med 240 tall (avdragene)
avdragsbetaling = np.array(avdrag240) #gjør liste til array.

x = np.arange(0, 240, 1) #definerer ønskede x verdier, i vårt tilfelle 240, grunnet 240 terminer.
y1 = nedbetaling_av_renter #definerer y1
y2 = avdragsbetaling #definerer y2
plt.stackplot(x, y2, y1, colors =['green', 'mediumseagreen'], labels = ["Avdrag", "Renter"]) #stackplot
plt.style.use('fivethirtyeight') #tema
plt.xlabel("Terminer (Måneder)") #x akse label
plt.ylabel("Månedsbeløp") #y akse label
plt.title("""Månedsbeløp for tilbakebetaling 
av lån på 1 million med 3% rente""") #tittel
plt.legend(loc='lower left',frameon=True) #legend posisjon
```
